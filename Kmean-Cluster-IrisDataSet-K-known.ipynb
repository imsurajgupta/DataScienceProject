```python
iris=load_iris()
```

```python
df=pd.DataFrame({
    'sepal_length':iris.data[:,0],
    'sepal_width':iris.data[:,1],
    'petal_length':iris.data[:,2],
    'petal_width':iris.data[:,3]
})
df.head()
```

```python
centroids={}
for i in range(3):
    result_list=[]
    result_list.append(df.loc[iris['target']==i]['sepal_length'].mean())
    result_list.append(df.loc[iris['target']==i]['sepal_width'].mean())
    centroids[i]=result_list
```

```python
centroids
```

```python
plt.scatter(df['sepal_length'],df['sepal_width'],c=iris.target,alpha=0.1) # alpha for opacity
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
colmap={0:'r',1:'b',2:'g'}

for i in range(3):
    plt.scatter(centroids[i][0],centroids[i][1],c=colmap[i])

plt.show()
```

```python
X=df.iloc[:,[0,1]].values
```

```python
from sklearn.cluster import KMeans

wcss=[]   # within cluster sum squar
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=100,c='red',label='Cluster1',marker="*")
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=100,c='blue',label='Cluster2')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=100,c='green',label='Cluster3')
```
